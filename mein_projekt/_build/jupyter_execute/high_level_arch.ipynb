## Frontend
Das Frontend unseres Projekts basiert auf dem Angular Framework und wird auf einer Linux-basierten virtuellen Maschine (VM) betrieben. <br>Die Entscheidung für Angular fiel aufgrund seiner vielfältigen integrierten Funktionen, die eine effiziente Verschlüsselung sowie eine optimale Interoperabilität mit diversen Backend-Frameworks ermöglichen. Die Verwendung einer Linux-VM gewährleistet eine stabile und zukunftssichere Betriebsumgebung. Zur Verarbeitung eingehender Anfragen ist die VM mit einem Apache Webserver ausgestattet, der die Anfragen nahtlos an das Backend weiterleitet.

## Backend
Das Backend unseres Systems wird durch das Django Framework realisiert, ein auf Python spezialisiertes Webframework.<br> Die Wahl fiel auf Django, da es im Vergleich zu Flask umfangreichere Funktionen bietet, insbesondere eingebaute Optionen für das Hosting von Machine-Learning-Modellen. Das Backend wird ebenfalls auf einer separaten Linux-VM gehostet, um eine klare Trennung vom Frontend zu gewährleisten. <br>Die Datenübertragung zwischen Frontend und Backend erfolgt vollständig verschlüsselt mittels HTTPS/TLS. Bei einer Risikobewertung, die eine Kreditkartenbetrugswahrscheinlichkeit von über 33% ergibt, werden die Daten zur weiteren Verarbeitung an die Datenbank weitergeleitet. Vor der Übermittlung in die Datenbank werden sicherheitsrelevante Daten tokenisiert.

## Datenbank
In unserer Datenbank werden sämtliche Daten tokenisiert gespeichert, wodurch sie ohne den entsprechenden Token wertlos sind. Ausgenommen davon sind bestimmte risikoarme Daten, die für weitere Verarbeitungsprozesse benötigt werden könnten. Als Datenbanksystem dient ein PostgreSQL Server, der eine stabile und sichere Umgebung für unsere Daten bietet.

## Rückmeldung
Im Falle eines existierenden Tokens in der Datenbank, welcher auf einen früheren Betrugsversuch hinweist, wird diese Information vom Datenbankserver an das Backend übermittelt.<br> Das Backend leitet diese Warnung anschließend an das Frontend weiter, wo sie dem Nutzer angezeigt wird. Dies stellt einen essenziellen Bestandteil unserer Sicherheitsarchitektur dar, um effektiv auf potenzielle Betrugsfälle reagieren zu können.
