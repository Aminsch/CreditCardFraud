# Frontend
Das Frontend basiert auf einem Angular Framework und läuft in einer Linux VM. Es wird Angular verwendet, weil es einige Built In Funktionen für die Verschlüsslung und die optimale Zusammenarbeit mit verschiedenen Backendframeworks vorhanden sind. Die Linux VM wird verwendet, weil es eine stabile Umgebung ist, die auch für die Zukunft gut geeignet ist. Die VM wird mit einem Apache Webserver ausgestattet, der die Anfragen an das Backend weiterleitet. 
# Backend
Das Backend basiert auf dem Django Framework, dieses ist ein Python optinmiertes Webframework. Es wird Django verwendet, weil es deutlich unmfangreicher als Flask ist und es built in Optionen für das Hosting von Machine Learning Modellen besitzt. Das Hosting des Backend findet ebenfalls auf einer Linux VM statt, wobei diese getrennt von der Frontend VM ist.
Die Verarbeitung von Daten zwischen Frontend und Backend findet komplett verschlüsselt statt, hierbei werden die Daten HTTPS/TLS verschlüsselt.
Falls die Risikobewertung ergibt, dass die Kreditkartenbetrugswahrscheinlichkeit höher als 1% ist, werden die Daten an die Datenbank weitergeleitet.
Wenn Daten weiterverarbeitet werden sollen, werden die sicherheitsrelevvante Daten Tokenisiert und erst dann in die Datenbank geschrieben.
# Datenbank
Die geschriebenen Daten in der Datenbank sind alle tokenisiert, sodass diese Informationen ohne den entsprechenden Token nutzlos sind, einige risikoarme Daten werden jedoch nicht tokenisiert, da diese für die weitere Verarbeitung benötigt werden könnten. Die Datenbank wird mit einem PostgreSQL Server realisiert, da dieser eine stabile und sichere Umgebung bietet.
# Rückmeldung
Falls ein Token bereits in der Datenbank existiert, also der dahinterstehende Name schoneinmal einen Fraud Alert erhalten hat, wird dies dem Backend vom Datenbank Server mitgeteilt.
Dieser Alert wird vom Backend an das Frontend weitergeleitet und dort dem Nutzer angezeigt.
